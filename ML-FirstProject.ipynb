```python
# Splitting data into two sets. 80 percent training and 20 percent test data
import pandas as pd
from sklearn.tree import DecisionTreeClassifier as dtc
# from --sklearn-- package ---model_selection-- module we are importing ---train_test_split--- function
from sklearn.model_selection import train_test_split
# To check accuracy of our module we use below import statement
from sklearn.metrics import accuracy_score


#train_test_split(X,y,test_size = 0.2) This function takes 3 arguments, X,y and test size would be 20 percent.
#This function returns a tuple with two values for X (trainX and test X) and two values for y (trainy and testy)
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size = 0.2)

#print (X_test)
model = dtc()
model.fit(X_train, y_train)
print((X_train))
print((X_test))
print('='*60)
prediction= model.predict(X_test)
#print(prediction)
score= accuracy_score(y_test,prediction)
score
```

```python
#Model Persistance : Training a model takes lot of time. So save a trained model and use it later.
import pandas as pd
from sklearn.tree import DecisionTreeClassifier as dtc
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import joblib # This joblib object used to save and loads models



X_train,X_test,y_train,y_test=train_test_split(X,y,test_size = 0.2)
model = dtc()
model.fit(X_train, y_train)
# dump/save the model now
joblib.dump(model, 'DTC_RecommenderSys_music.joblib')


```
